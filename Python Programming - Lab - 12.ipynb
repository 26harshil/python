```python
#import matplotlib below
import matplotlib.pyplot as plt
```

```python


# Data
x = range(1, 11)
y = [1, 5, 9, 7, 5, 6, 3, 2, 4, 9]

# Plotting
plt.plot(x, y, label='Line')

# Labels and title
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("x vs y")

# Grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph


# Plotting
plt.plot(x,cxMarks,label="cx",marker='o',color='b',ls=':')
plt.plot(x,cyMarks,label="cys",marker='<',color='r',ls=':')

# Labels and title
plt.xlabel("x - axis")
plt.text(7,0.8,"lowest point")
plt.ylabel("y - axis")
plt.title("x vs y")

# Grid and legend

plt.legend()

# Display the plot
plt.show()
```

### 04) WAP to demonstrate the use of Pie chart.

### 05) WAP to demonstrate the use of Bar chart.

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
plt.bar(fruits,sales,color='k')

plt.title("sales vs fruits")
plt.xlabel("fruits")
plt.ylabel("sales")
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = np.random.randn(100)
 
sns.histplot(data, bins=30, kde=True, color='lightgreen', edgecolor='red')
plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
plt.bar(fruits,sales,color='b')
for i in range(len(fruits)):
    plt.text(i,sales[i]+0.2,str(sales[i]))
plt.title("sales vs fruits")
plt.xlabel("fruits")
plt.ylabel("sales")
plt.show()
```
